```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
```

```python
img = load_img('image.19.jpeg',target_size=(200,200)) 
```

```python
img
```

```python
datagen = ImageDataGenerator(
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        brightness_range = (0.5, 1.5))
```

```python
X = img_to_array(img)
```

```python
X.shape
```

```python
input_batch = X.reshape(1,200,200,3)
```

```python
i = 0
for batch in datagen.flow(input_batch, batch_size=1,save_to_dir='preview'):
    i += 1
    if i > 20:
        break  
```
